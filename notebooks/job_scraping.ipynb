```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import re
```

```python
job = "data scientist"
page = 0
job_formatted = re.sub(" ", "%20", job)
main_url = f"https://www.mycareersfuture.gov.sg/search?search={job_formatted}&sortBy=new_posting_date&page={page}"
# main_css = """.bg-white.mb3.w-100.dib.v-top.pa3.no-underline.flex-ns.flex-wrap.JobCard__card___22xP3"""
title_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[1]/section/div[2]/div[1]/span"""
company_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[1]/section/div[2]/p"""
location_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[2]/section/p[1]"""
employment_type_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[2]/section/p[2]"""
profession_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[2]/section/p[3]"""
category_xpath = """//*[@id="job-card-{}"]/div/a/div[1]/div[2]/section/p[4]"""
all_xpath = """//*[@id="job-card-{}"]/div/a"""
```

```python
options = webdriver.EdgeOptions()
options.add_argument("headless")
driver = webdriver.Edge("../msedgedriver.exe", options=options)
driver.get(main_url)
delay = 20  # seconds
# all_info = driver.find_elements(by=By.CLASS_NAME, value=main_class)
for i in range(22):
    try: 
        element_present = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, all_xpath.format(i))))
    except TimeoutException:
        print("Loading took too much time!")
    all_info = driver.find_elements(by=By.XPATH, value=all_xpath.format(i))
    for info in all_info:
        title = info.find_element(by=By.XPATH, value=title_xpath.format(i)).text
        company = info.find_element(by=By.XPATH, value=company_xpath.format(i)).text
        location = info.find_element(by=By.XPATH, value=location_xpath.format(i)).text
        employment_type = info.find_element(
            by=By.XPATH, value=employment_type_xpath.format(i)
        ).text
        profession = info.find_element(by=By.XPATH, value=profession_xpath.format(i)).text
        category = info.find_element(by=By.XPATH, value=category_xpath.format(i)).text
        print(title)
        print(company)
        print(location)
        print(employment_type)
        print(profession)
        print(category)
        print()
```
